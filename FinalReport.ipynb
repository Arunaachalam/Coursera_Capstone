# Methodology

To understand the locations and clusters of available shops, first the cleaning of data is done for the data from wiki. The data are sorted with the Postal Code to easily append the geolocation data in the next step.

The geolocatin data is loaded, which is based on the postal code. This data is appended to the previous dataframe, to derive a new dataframe which consists of Postal Code, Borough, Neighborhood, Latitude, and Longitude data.

The data are sorted and grouped through Boroughs, as it is an easy way of visualizing datas. The average value of the geolocation for each borough is chosen as the center geolocation of the borough.

With the center geolocation data, with a reasonable limit of 100 and radius of 1000, Foursome API is used to extract neighborhood data. The data is cleaned and sorted for each borough and are appended to a single data frame which contains the details of establishment name, geolocation, and categories. This data is plotted and analysed to see the possibility of population clusters based on the number of attractions available

# Discussion

From the results, the clusters are determined based on the number of shops and the transport system will be designed especially for shopping district. From the Foursquare data, it is clear that,

- the profitable place or the shopping district is in the central areas of toronto, especially the boroughs Central Toronto, Downtown Toronto, West Toronto, and East Toronto. Apart from these areas Mississauga has more shops, but as it is situated outside the city near the Airport, short distance transits are not adviced.

- The Future transport construction should be concentrated in the boroughs of Central Toronto, DownTown Toronto, West Toronto, East Toronto and East York.

- Though East York does not have any importance, due to its close proximity to the shopping district, there is a scope for future development

- York and North York are near airports, so future transport construction is meaningless in these areas

- Sarborough is a resdential area and not a shopping district

# Conclusion

In this project, the places where the new future transportation system can be set up with profits are determined. The data are obtained from wiki and other sources like Foursquare and are processed to obtain the geolocation data with more shops.

On deeper analysis on the number of shops and proximity the Boroughs of Central Toronto, Downtown Toronto, West Toronto, East Toronto and East York are chosen for the future city project.
